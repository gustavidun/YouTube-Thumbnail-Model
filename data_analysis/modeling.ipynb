```python
import pandas as pd

df = pd.read_pickle("data/videos_labeled.pkl")
df
```

```python
import statsmodels.api as sm

import statsmodels.formula.api as smf

md = smf.mixedlm("day19 ~ subscribers + conflict + text + juxtaposition + cliffhanger + arrows + monochrony", df, groups=df["channel_title"])

mdf = md.fit()

print(mdf.summary())

```
