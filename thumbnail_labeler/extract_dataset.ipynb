### Ground Truth Dataset Extraction
We extract a ground truth dataset for manual labeling by sampling randomly from the full dataset. 

```python
#filter language
from lingua import LanguageDetectorBuilder, Language
import pandas as pd

df_non_trimmed = pd.read_pickle("videos.pkl")
detector = LanguageDetectorBuilder.from_all_languages().build()

drop_list = []

for i, title in enumerate(df_non_trimmed["title"].to_list()):
    print("Processing ", i)
    conf = detector.compute_language_confidence_values(title)
    if conf[0].language != Language.ENGLISH:
        drop_list.append(i)

df = df_non_trimmed.drop(df_non_trimmed.index[drop_list])

#remove blacklisted videos
df = df[df["blacklisted"] != True]

print(len(df))
print(len(df_non_trimmed))
```

```python


sampled_df = df.sample(n = 300, ignore_index=True)
sampled_df_len = len(sampled_df)

new_df = pd.DataFrame(
    {
        "url": sampled_df["thumbnail_url"],
        "title": sampled_df["title"],
        "id": sampled_df["video_id"],
        "question": [False]*sampled_df_len,
        "text": [False]*sampled_df_len,
        "faces": ["none"]*sampled_df_len,
        "arrows": [False]*sampled_df_len,
        "monochrony": [False]*sampled_df_len,
        "juxtaposition": [False]*sampled_df_len,
        "cliffhanger": [False]*sampled_df_len,
        "conflict": [False]*sampled_df_len,
        "reviewed": [False]*sampled_df_len,
    }
)

new_df.to_pickle("data.pkl")

new_df[:5]
```
